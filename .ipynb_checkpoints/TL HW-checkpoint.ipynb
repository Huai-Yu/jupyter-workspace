# 載入套件及資料集

```python
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten

from tensorflow.keras.layers import Conv2D, MaxPool2D, GlobalAveragePooling2D

# #optimizer:
from tensorflow.keras.optimizers import Adam

from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical

import os

os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

# 處理資料
## A資料集使用MNIST，類別為0~9的數字，每張圖的尺寸為  28×28

```python
# Load data
(X_train, y_train0), (X_test, y_test0) = datasets.mnist.load_data()
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train0.shape)
print("y_test:", y_test0.shape)
```

```python
idx = np.random.randint(X_train.shape[0])
X_sample = X_train[idx]
y_sample = y_train0[idx]
plt.imshow(X_sample)
plt.title(y_sample)
plt.axis('off');
```

```python
X_train = X_train.reshape(60000, 28, 28, 1)
X_test = X_test.reshape(10000, 28, 28, 1)
```

```python
# Normalize the range of featurs
X_train = X_train / X_train.max()
X_test = X_test / X_test.max()

# One-hot encoding
y_train = to_categorical(y_train0, 10)
y_test = to_categorical(y_test0, 10)
```

# 建立分類MNIST的神經網路

```python
CNN_layers = [Conv2D(28, (3, 3), input_shape=(28, 28, 1), padding='same', activation='relu'),
        MaxPool2D(),
        Conv2D(64, (3, 3), padding='same', activation='relu'),
        MaxPool2D(),
        Conv2D(128, (3, 3), padding='same', activation='relu'),
        GlobalAveragePooling2D()]

FC_layers = [Dense(units=256, activation='relu'),
       Dense(units=10, activation='softmax')]
```

```python
model = Sequential(CNN_layers+FC_layers)
model.summary()
```

```python
model = Sequential()

#First convolutional block
model.add(Conv2D(28, (3, 3), input_shape=(28, 28, 1), padding='same', activation='relu'))
model.add(MaxPool2D())

#Second convolutional block
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(MaxPool2D())

#Third convolutional block
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(GlobalAveragePooling2D())

#Fully-connected layers as a classfier
model.add(Dense(units=256, activation='relu'))

#Ouput layer: # of neurons = # of classes with softmax activation
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(),metrics=['categorical_accuracy'])
```

# 訓練模型

```python
model.fit(X_train, y_train, batch_size=128, epochs=3,validation_data=(X_test, y_test))
```

```python
model.save_weights('LeNet5_MNIST.h5')
```

# 模型預測結果

```python
model.load_weights('LeNet5_MNIST.h5')

score_train = model.evaluate(X_train, y_train)
score_test = model.evaluate(X_test, y_test)

print(f'Train Accuracy: {score_train[1]*100}')
print(f'Test Accuracy: {score_test[1]*100}')
```
